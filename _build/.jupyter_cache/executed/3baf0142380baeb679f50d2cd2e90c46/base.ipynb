```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([(1,2), (1,3), (2,3), (1,4), (2,4), (3,3), (3,5)])
nx.draw_circular(G, with_labels=True, font_weight="bold", font_size=20, node_size=600)
```

```python
[n for n in G.nodes]
```

```python
[e for e in G.edges]
```
