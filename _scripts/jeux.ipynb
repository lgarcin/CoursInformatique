```python
from collections import defaultdict

# Définition du graphe biparti pour un morpion 2x2
graphe = defaultdict(list)

# État initial (grille vide)
etat_initial = (".", ".", ".", ".")

# Génération des états après un coup de J1 (X)
etats_j1 = [
    ("X", ".", ".", "."), (".", "X", ".", "."), 
    (".", ".", "X", "."), (".", ".", ".", "X")
]
graphe[etat_initial] = etats_j1

# Génération des états après un coup de J2 (O) à partir des états de J1
for etat in etats_j1:
    for i in range(4):
        if etat[i] == ".":
            nouvel_etat = list(etat)
            nouvel_etat[i] = "O"
            graphe[etat].append(tuple(nouvel_etat))

# Affichage du graphe
for etat, voisins in graphe.items():
    print(f"{etat} → {voisins}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Création du graphe orienté
G = nx.DiGraph()

# Ajout des sommets et arêtes
for etat, voisins in graphe.items():
    for voisin in voisins:
        G.add_edge(etat, voisin)

# Dessin du graphe
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)  # Position des sommets
nx.draw(G, pos, with_labels=True, node_size=3000, font_size=10, font_color="white", node_color="blue")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# États du jeu : grilles 2x2
etat_initial = (".", ".", ".", ".")
etats_j1 = [
    ("X", ".", ".", "."), (".", "X", ".", "."), 
    (".", ".", "X", "."), (".", ".", ".", "X")
]

# Génération du graphe biparti
graphe = {etat_initial: etats_j1}
for etat in etats_j1:
    graphe[etat] = []
    for i in range(4):
        if etat[i] == ".":
            nouvel_etat = list(etat)
            nouvel_etat[i] = "O"
            graphe[etat].append(tuple(nouvel_etat))

# Création du graphe orienté
G = nx.DiGraph()
for etat, voisins in graphe.items():
    for voisin in voisins:
        G.add_edge(etat, voisin)

# Fonction pour formater un état sous forme de grille
def format_grille(etat):
    return f"{etat[0]} {etat[1]}\n{etat[2]} {etat[3]}"

# Position des nœuds
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)  # Générer une disposition
labels = {etat: format_grille(etat) for etat in G.nodes()}  # Grilles au lieu de tuples

# Dessin du graphe
nx.draw(G, pos, with_labels=True, labels=labels, node_size=3000, 
        font_size=10, font_color="black", node_color="lightblue", edge_color="gray")

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# États du jeu : grilles 2x2
etat_initial = (".", ".", ".", ".")
etats_j1 = [
    ("X", ".", ".", "."), (".", "X", ".", "."), 
    (".", ".", "X", "."), (".", ".", ".", "X")
]

# Génération du graphe biparti
graphe = {etat_initial: etats_j1}
for etat in etats_j1:
    graphe[etat] = []
    for i in range(4):
        if etat[i] == ".":
            nouvel_etat = list(etat)
            nouvel_etat[i] = "O"
            graphe[etat].append(tuple(nouvel_etat))

# Création du graphe orienté
G = nx.DiGraph()
for etat, voisins in graphe.items():
    for voisin in voisins:
        G.add_edge(etat, voisin)

# Fonction pour formater un état sous forme de vraie grille
def dessiner_grille(ax, etat):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)

    # Tracer la grille
    ax.plot([1, 1], [0, 2], 'k', lw=2)  # Barre verticale
    ax.plot([0, 2], [1, 1], 'k', lw=2)  # Barre horizontale

    # Ajouter les symboles X et O
    positions = [(0.5, 1.5), (1.5, 1.5), (0.5, 0.5), (1.5, 0.5)]
    for i, (x, y) in enumerate(positions):
        if etat[i] != ".":
            ax.text(x, y, etat[i], fontsize=16, ha='center', va='center', fontweight='bold')

# Dessin du graphe avec les grilles dans les nœuds
fig, ax = plt.subplots(figsize=(10, 6))
pos = nx.spring_layout(G)  # Générer une disposition

# Dessiner les arêtes
nx.draw(G, pos, with_labels=False, edge_color="gray", arrows=True, node_size=0)  

# Dessiner chaque état sous forme de mini-grille
for etat, (x, y) in pos.items():
    ax_inset = fig.add_axes([x * 0.6 + 0.2, y * 0.6 + 0.2, 0.12, 0.12])  # Positionner chaque grille
    dessiner_grille(ax_inset, etat)

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# États du jeu : grilles 2x2
etat_initial = (".", ".", ".", ".")
etats_j1 = [
    ("X", ".", ".", "."), (".", "X", ".", "."), 
    (".", ".", "X", "."), (".", ".", ".", "X")
]

# Génération du graphe biparti
graphe = {etat_initial: etats_j1}
for etat in etats_j1:
    graphe[etat] = []
    for i in range(4):
        if etat[i] == ".":
            nouvel_etat = list(etat)
            nouvel_etat[i] = "O"
            graphe[etat].append(tuple(nouvel_etat))

# Création du graphe orienté
G = nx.DiGraph()
for etat, voisins in graphe.items():
    for voisin in voisins:
        G.add_edge(etat, voisin)

# Position des nœuds
pos = nx.spring_layout(G)

# Dessin des arêtes du graphe
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=False, edge_color="gray", arrows=True, node_size=0)  

# Fonction pour dessiner une grille 2x2 avec des X et O dans un nœud
def dessiner_grille(ax, etat):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)
    ax.axis("off")

    # Tracer la grille
    ax.plot([1, 1], [0, 2], 'k', lw=2)  # Barre verticale
    ax.plot([0, 2], [1, 1], 'k', lw=2)  # Barre horizontale

    # Ajouter les X et O dans la grille
    positions = [(0.5, 1.5), (1.5, 1.5), (0.5, 0.5), (1.5, 0.5)]
    for i, (x, y) in enumerate(positions):
        if etat[i] != ".":
            ax.text(x, y, etat[i], fontsize=16, ha='center', va='center', fontweight='bold')

# Ajouter une mini-grille à chaque nœud
fig, ax = plt.subplots(figsize=(10, 6))
nx.draw(G, pos, with_labels=False, edge_color="gray", arrows=True, node_size=0)  

for etat, (x, y) in pos.items():
    ax_inset = fig.add_axes([x * 0.6 + 0.2, y * 0.6 + 0.2, 0.12, 0.12])  # Ajustement pour placer les grilles dans les nœuds
    dessiner_grille(ax_inset, etat)

plt.show()
```
